## Benchmarking Prime Generation Times

```python
k=[10000, 50000, 100000, 125000, 150000] # Number of primes
p_e =[1020, 25800, 98000, 169000, 242000] # Time for Sieve of Eratosthenes
p_c = [6.59, 59.9, 112, 151, 180] # Time for Croft Spiral Sieve
p_s = [1.8, 24.3, 51, 67.7, 82.7] # Time for sympy primerange
plt.plot(k, p_c, label="Croft", color="b")
plt.plot(k, p_s, label="sympy", color="g")
plt.legend()
plt.xlabel("Number of Primes generated")
plt.ylabel("Time in ms")
plt.figure()

plt.plot(k, p_e, label="Eratosthenes", color="r")
plt.plot(k, p_c, label="Croft", color="b")
plt.legend()
plt.xlabel("Number of Primes generated")
plt.ylabel("Time in ms")
plt.figure()


plt.plot(k, [i / j for i, j in zip(p_c, p_s)], label="Croft v/s sympy", color="r")
plt.legend()
plt.xlabel("Number of Primes generated")
plt.ylabel("Multiplier in Time")
plt.figure()
```

## Algorithm Benchmark

```python
k=[1000, 5000, 10000, 15000, 16384] # Number of primes
t_a1=[1.93, 290, 2412, 8478, 10849] # Time for Algorithm-1
t_a1_t=[1.86, 286, 2341, 8077, 10418.9] # Time for Algorithm-1 with Turbo'ed CPU
t_a2=[0.0607, 2.51, 13.5, 34, 45.2] # Time for Algorithm-2
t_a2_t=[0.0592, 2.34, 12.9, 32.8, 38.9] # Time for Algorithm-2 with Turbo'ed CPU

plt.plot(k, t_a1, label="Algo1", color="b")
plt.plot(k, t_a2, label="Algo2", color="g")
plt.legend()
plt.xlabel("Number of Primes Used.")
plt.ylabel("Time in s")
plt.title("Comparision between Two Algorithms")
plt.figure()

plt.plot(k, [i / j for i, j in zip(t_a1, t_a1_t)], label="Algo1", color="b")
plt.plot(k, [i / j for i, j in zip(t_a2, t_a2_t)], label="Algo2", color="g")
plt.legend()
plt.xlabel("Number of Primes Used.")
plt.ylabel("Time in s")
plt.title("Results of Turbo-ing CPU")
plt.figure()
```
